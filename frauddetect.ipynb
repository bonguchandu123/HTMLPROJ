```python
%%writefile app.py
import streamlit as st
import numpy as _onp
import pandas as pd
import time
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, f1_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt

st.set_page_config(page_title="üîí Quantum + Classical Fraud Detection", layout="wide")

# =========================
# Quantum setup
# =========================
n_qubits = 4
dev = qml.device("default.qubit", wires=n_qubits)

def feature_map(x):
    for i in range(n_qubits):
        qml.RY(x[i], wires=i)
    for i in range(n_qubits-1):
        qml.CNOT(wires=[i, i+1])

def ansatz(weights):
    for i in range(n_qubits):
        qml.RY(weights[i], wires=i)
        qml.RZ(weights[i+n_qubits], wires=i)

@qml.qnode(dev, interface="autograd")
def circuit(x, weights):
    feature_map(x)
    ansatz(weights)
    return qml.expval(qml.PauliZ(0))

def predict_batch(W, Xb):
    outs = []
    for x in Xb:
        z = circuit(x, W)
        p = 0.5 * (1 + z)
        outs.append(np.clip(p, 1e-7, 1-1e-7))
    return np.stack(outs)

def bce_loss(W, Xb, yb):
    p = predict_batch(W, Xb)
    return -np.mean(yb*np.log(p) + (1-yb)*np.log(1-p))

# =========================
# Toy & CSV utilities
# =========================
def make_toy_fraud(n=1200, seed=42):
    rng = _onp.random.default_rng(seed)
    n0 = int(n*0.9)
    X0 = rng.normal(loc=[0.2,0.4,0.3,0.2], scale=[0.1,0.15,0.15,0.1], size=(n0,4))
    y0 = _onp.zeros(n0)
    n1 = n - n0
    X1 = rng.normal(loc=[0.7,0.8,0.8,0.7], scale=[0.15,0.15,0.1,0.1], size=(n1,4))
    y1 = _onp.ones(n1)
    X = _onp.vstack([X0, X1])
    y = _onp.concatenate([y0, y1])
    X = _onp.clip(X, 0, 1)
    return X, y.astype(int)

def parse_uploaded_csv(df):
    cols = {c.lower(): c for c in df.columns}
    required = ["amount","time","device","merchant_risk","label","merchant_category","transaction_type","cardholder_age"]
    for c in required:
        if c not in cols:
            raise ValueError(f"CSV missing column: {c}")
    dmap = {"mobile":0.2,"desktop":0.5,"atm":0.8}
    device = _onp.array([dmap.get(s.lower(),0.5) for s in df[cols["device"]].astype(str)])
    cat_enc = LabelEncoder()
    merchant_cat = cat_enc.fit_transform(df[cols["merchant_category"]])
    merchant_cat = merchant_cat / max(1, merchant_cat.max())
    tmap = {"online":0.2,"in-person":0.5,"atm":0.8}
    trans_type = _onp.array([tmap.get(s.lower(),0.5) for s in df[cols["transaction_type"]].astype(str)])
    amount = _onp.clip(df[cols["amount"]].astype(float)/1000.0,0,1)
    time_arr = _onp.clip(df[cols["time"]].astype(float)/24.0,0,1)
    risk = _onp.clip(df[cols["merchant_risk"]].astype(float),0,1)
    age = _onp.clip(df[cols["cardholder_age"]].astype(float)/100.0,0,1)
    label = df[cols["label"]].astype(int)
    X = _onp.stack([amount,time_arr,device,risk],axis=1)
    X_extra = _onp.stack([merchant_cat, trans_type, age],axis=1)
    return X, X_extra, label

# =========================
# Training
# =========================
@st.cache_resource(show_spinner=True)
def train_quantum_classifier(X_in=None, y_in=None, epochs=20, batch_size=64, stepsize=0.2, seed=123):
    if X_in is None or y_in is None:
        X, y = make_toy_fraud(n=1400, seed=seed)
    else:
        X, y = X_in, y_in
    scaler = StandardScaler()
    X = scaler.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed, stratify=y)
    np.random.seed(seed)
    W = np.array(0.01 * np.random.randn(2*n_qubits), requires_grad=True)
    opt = qml.GradientDescentOptimizer(stepsize=stepsize)
    X_train_q = np.array(X_train, requires_grad=False)
    y_train_q = np.array(y_train, requires_grad=False)
    for epoch in range(epochs):
        idx = _onp.random.choice(len(X_train), min(batch_size,len(X_train)), replace=False)
        batchX = X_train_q[idx]
        batchY = y_train_q[idx]
        W = opt.step(lambda w: bce_loss(w,batchX,batchY), W)
    probs = predict_batch(W, np.array(X_test))
    preds = (probs > 0.5).astype(int)
    auc = roc_auc_score(y_test, _onp.asarray(probs))
    f1 = f1_score(y_test, _onp.asarray(preds))
    return W, scaler, (X_test, y_test, probs, preds, auc, f1)

@st.cache_resource(show_spinner=True)
def train_classical_baseline(X_in=None, X_extra_in=None, y_in=None):
    if X_in is None or y_in is None:
        X, y = make_toy_fraud(n=1400)
        X_extra = _onp.zeros((len(X),3))
    else:
        X, X_extra, y = X_in, X_extra_in, y_in
    X_all = _onp.hstack([X,X_extra])
    scaler = StandardScaler()
    X_all_scaled = scaler.fit_transform(X_all)
    X_train, X_test, y_train, y_test = train_test_split(X_all_scaled, y, test_size=0.2, stratify=y)
    clf_rf = RandomForestClassifier(n_estimators=100, random_state=42)
    clf_rf.fit(X_train, y_train)
    clf_lr = LogisticRegression().fit(X_train, y_train)
    return scaler, clf_rf, clf_lr, X_test, y_test

# =========================
# UI
# =========================
st.title("üîí Quantum + Classical Fraud Detection")
st.markdown("Upload CSV or use toy data to compare Quantum, Classical, and Hybrid predictions with extra credit card features.")

with st.expander("üì• Upload CSV (optional)"):
    st.write("Required columns: amount, time, device, merchant_risk, label, merchant_category, transaction_type, cardholder_age")
    up = st.file_uploader("Upload CSV", type=["csv"])
    X_custom = None
    X_extra_custom = None
    y_custom = None
    if up is not None:
        try:
            df = pd.read_csv(up)
            X_custom, X_extra_custom, y_custom = parse_uploaded_csv(df)
            st.success(f"Loaded {len(df)} rows")
            st.dataframe(df.head())
        except Exception as e:
            st.error(f"CSV parse error: {e}")

# Sidebar
st.sidebar.header("Mode")
mode = st.sidebar.selectbox("Select mode", ["Pre-trained demo", "Train Quantum Model"])
colL, colR = st.columns([1,1])

epochs = 20
batch_size = 64
stepsize = 0.2
seed = 123

if mode=="Train Quantum Model":
    with colL:
        st.subheader("‚öôÔ∏è Training Settings")
        epochs = st.slider("Epochs", 5, 50, 20)
        batch_size = st.slider("Batch size", 16, 256, 64)
        stepsize = st.slider("Learning rate", 0.05, 0.6, 0.2)
        seed = st.number_input("Random seed", 0, 9999, 123)
        if st.button("üöÄ Train Quantum"):
            W, scaler, eval_pack = train_quantum_classifier(X_in=X_custom, y_in=y_custom,
                                                            epochs=epochs, batch_size=batch_size,
                                                            stepsize=stepsize, seed=seed)
            st.success("Quantum training finished!")
        else:
            W, scaler, eval_pack = train_quantum_classifier(epochs=5, X_in=X_custom, y_in=y_custom)
else:
    st.info("Using pre-trained demo model (toy data).")
    W, scaler, eval_pack = train_quantum_classifier(epochs=5, X_in=X_custom, y_in=y_custom)

# Classical baseline
scaler_clf, clf_rf, clf_lr, X_test_cls, y_test_cls = train_classical_baseline(X_in=X_custom, X_extra_in=X_extra_custom, y_in=y_custom)

# Quantum evaluation
X_test, y_test, probs, preds, auc, f1 = eval_pack
with colR:
    st.subheader("üìä Quantum Model Performance")
    st.write(f"AUC: {auc:.3f}, F1: {f1:.3f}")
    st.text("Classification report:")
    st.code(classification_report(y_test, _onp.asarray(preds), digits=3))
    st.text("Confusion matrix:")
    st.write(confusion_matrix(y_test, _onp.asarray(preds)))

# Classical comparison
st.subheader("üìà Classical Baseline Comparison")
st.write("Random Forest + Logistic Regression on same dataset:")
probs_rf = clf_rf.predict_proba(X_test_cls)[:,1]
preds_rf = clf_rf.predict(X_test_cls)
probs_lr = clf_lr.predict_proba(X_test_cls)[:,1]
preds_lr = clf_lr.predict(X_test_cls)
st.write(f"Random Forest: AUC={roc_auc_score(y_test_cls, probs_rf):.3f}, F1={f1_score(y_test_cls, preds_rf):.3f}")
st.write(f"Logistic Regression: AUC={roc_auc_score(y_test_cls, probs_lr):.3f}, F1={f1_score(y_test_cls, preds_lr):.3f}")



# =========================
# Inference panel
# =========================
st.markdown("---")
st.subheader("üß™ Try a Transaction (Credit Card Extra Features)")

with st.sidebar:
    st.header("Transaction Input")
    amount = st.number_input("Amount ($)", 1.0, 5000.0, 135.0)
    hour = st.slider("Hour of day", 0, 23, 2)
    device = st.selectbox("Device", ["Mobile","Desktop","ATM"], index=0)
    merchant_risk = st.slider("Merchant risk (0=low,1=high)", 0.0, 1.0, 0.65, 0.01)
    merchant_category = st.selectbox("Merchant Category", ["Electronics","Grocery","Entertainment"], index=0)
    transaction_type = st.selectbox("Transaction Type", ["Online","In-Person","ATM"], index=0)
    cardholder_age = st.number_input("Cardholder Age", 18, 100, 35)

device_map = {"Mobile":0.2,"Desktop":0.5,"ATM":0.8}
cat_map = {"Electronics":0.0,"Grocery":0.5,"Entertainment":1.0}
type_map = {"Online":0.2,"In-Person":0.5,"ATM":0.8}

x_raw = _onp.array([amount/1000.0, hour/24.0, device_map[device], merchant_risk])
x_extra = _onp.array([cat_map[merchant_category], type_map[transaction_type], cardholder_age/100.0])
x_scaled = scaler.transform([x_raw])
p_q = float(predict_batch(W, np.array(x_scaled))[0])
st.subheader("üö® Quantum Prediction")
if p_q>0.5: st.error(f"‚ö†Ô∏è Fraud likely (p={p_q:.2f})")
else: st.success(f"‚úÖ Legitimate (p={p_q:.2f})")
x_cls_scaled = scaler_clf.transform([_onp.hstack([x_raw,x_extra])])
p_rf = clf_rf.predict_proba(x_cls_scaled)[:,1][0]
p_lr = clf_lr.predict_proba(x_cls_scaled)[:,1][0]
st.subheader("üìä Classical Predictions")
st.write(f"Random Forest: p={p_rf:.2f}")
st.write(f"Logistic Regression: p={p_lr:.2f}")
p_hybrid = (p_q + p_rf)/2
st.subheader("‚ö° Hybrid Prediction (Quantum + Random Forest)")
if p_hybrid>0.5: st.error(f"‚ö†Ô∏è Fraud likely (p={p_hybrid:.2f})")
else: st.success(f"‚úÖ Legitimate (p={p_hybrid:.2f})")

# Random transaction simulation
st.markdown("---")
st.subheader("üéÆ Simulate Random Transactions")
simulate_btn = st.button("Run Simulation")
if simulate_btn:
    sim_rows = []
    for i in range(10):
        amt = _onp.random.randint(10, 2000)
        hr = _onp.random.randint(0,23)
        dev_name = _onp.random.choice(["Mobile","Desktop","ATM"])
        risk = _onp.random.rand()
        cat = _onp.random.choice(["Electronics","Grocery","Entertainment"])
        ttype = _onp.random.choice(["Online","In-Person","ATM"])
        age = _onp.random.randint(18,80)
        x_sim = _onp.array([amt/1000.0, hr/24.0, device_map[dev_name], risk])
        x_extra_sim = _onp.array([cat_map[cat], type_map[ttype], age/100.0])
        x_sim_scaled = scaler.transform([x_sim])
        p_q_sim = float(predict_batch(W, np.array(x_sim_scaled))[0])
        x_cls_scaled_sim = scaler_clf.transform([_onp.hstack([x_sim,x_extra_sim])])
        p_rf_sim = clf_rf.predict_proba(x_cls_scaled_sim)[:,1][0]
        p_hybrid_sim = (p_q_sim+p_rf_sim)/2
        sim_rows.append([amt,hr,dev_name,risk,cat,ttype,age,p_q_sim,p_rf_sim,p_hybrid_sim])
        time.sleep(0.2)
    sim_df = pd.DataFrame(sim_rows,columns=["Amount","Hour","Device","Risk","MerchantCat","TransType","Age","Quantum","RF","Hybrid"])
    st.dataframe(sim_df)

st.caption("Hackathon-ready Quantum + Classical + Hybrid Fraud Detection with credit card features and visual explainability.")




```
